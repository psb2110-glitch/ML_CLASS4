```python
x=[1,2,3,4,5]
y=[3,6,2,4,5]
p=int(input("Enter the value for x: "))
sum_x=sum(x)
sum_y=sum(y)
mean_x=sum_x/len(x)
mean_y=sum_y/len(y)
a=0
b=0
for i in range(len(x)):
    a=a+((x[i]-mean_x)*(y[i]-mean_y))
    b=b+((x[i]-mean_x)**2)

m=a/b
print("The Slope is:",m)
c=(sum_y-(m*sum_y))/len(x)
print("The Intercept is:",c)
x_given=p
y_new=((m*x_given)+c)
print("The Predicted value of y for given x is:",y_new)
```
